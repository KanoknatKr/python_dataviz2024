## filter เอาข้อมูล comments ที่มียอดไลค์มากกกว่า 400

```python
doctor_df_filtered = doctor_df[doctor_df['ยอดไลค์'] > 400]
doctor_df_filtered
```

## filter เอาข้อมูล comments ที่มีจำนวนการตอบกลับมากที่สุด

```python
doctor_df_sorted = doctor_df.sort_values('จำนวนการตอบกลับ', ascending=False)
doctor_df_top5 = doctor_df_sorted.head(5)
doctor_df_top5
```

## filter เอาข้อมูล comments ที่ยาวที่สุด

```python
doctor_df['comment_length'] = doctor_df['comments'].str.len()
doctor_df_sorted = doctor_df.sort_values('comment_length', ascending=False)
top_5_comments = doctor_df_sorted.head(5)
top_5_comments
```

```python
# prompt: filter เอาข้อมูลที่ดีที่สุด โดนเฉลี่ยจาก จำนวน comment_lengthที่ยาวที่สุด ยอดไลค์มากที่สุด และจำนวนการตอบกลับมากที่สุด 5 ข้อมูล

# คำนวณค่าเฉลี่ยของ comment_length, ยอดไลค์, และจำนวนการตอบกลับ
doctor_df['average_score'] = (doctor_df['comment_length'] + doctor_df['ยอดไลค์'] + doctor_df['จำนวนการตอบกลับ']) / 3

# เรียงลำดับข้อมูลตาม average_score จากมากไปน้อย
doctor_df_sorted = doctor_df.sort_values('average_score', ascending=False)

# เลือก 5 ข้อมูลที่ดีที่สุด
top_5_average = doctor_df_sorted.head(5)
top_5_average
```

## เลือกนำเสนอข้อมูล 2 จุดที่น่าสนใจจากจำนวน comment_length ที่มากที่สุด

```python
doctor_df['comment_length'] = doctor_df['comments'].str.len()
doctor_df_sorted = doctor_df.sort_values('comment_length', ascending=False)
top_2_comments = doctor_df_sorted.head(2)
top_2_comments
```
