```python
import pandas as pd
import os
from google.colab import drive

# เชื่อมต่อ Google Drive
drive.mount('/content/drive')

# กำหนดเส้นทางไปยังไฟล์ Excel
path = '/content/drive/MyDrive/data_viz_2024_DATA/test-TidyData_2024.xlsx'

# อ่านไฟล์ Excel โดยใช้ pd.read_excel
df = pd.read_excel(path, engine='openpyxl')  # ใช้เอ็นจิ้นที่เหมาะสมสำหรับไฟล์ Excel

# ทำการ split ข้อมูลในคอลัมน์ month ออกให้เหลือแค่ตัวเลขปี
df[['month', 'year']] = df['month'].str.split('_', expand=True)
df['year'] = df['year'].astype(int)  # แปลงข้อมูลในคอลัมน์ 'year' เป็นตัวเลข

# คำนวณค่าเฉลี่ยของ 'Chlo_a หรือ ความเข้มข้นของคลอโรฟิลล์ a' สำหรับแต่ละปี
mean_chlo_a_per_year = df.groupby('year')['Chlo_a'].mean()

# แสดงค่าเฉลี่ยของ 'Chlo_a' สำหรับแต่ละปี
print(mean_chlo_a_per_year)

# คำนวณความแตกต่างของค่าเฉลี่ยระหว่างปี 2018 และ 2019
chlo_a_difference = mean_chlo_a_per_year[2018] - mean_chlo_a_per_year[2019]
print("ความแตกต่างของค่าเฉลี่ย 'Chlo_a' ระหว่างปี 2018 และ 2019 = ", chlo_a_difference)
```

# 2. สร้างตารางค่าเฉลี่ย Chlo_b ทุกๆ station ของแต่ละเดือน

```python
# สร้าง mapping ของชื่อเดือนเป็นหมายเลขเดือน
month_mapping = {
    'Jan': 1, 'Feb': 2, 'Mar': 3, 'Apr': 4,
    'May': 5, 'June': 6, 'July': 7, 'Aug': 8,
    'Sep': 9, 'Oct': 10, 'Nov': 11, 'Dec': 12
}

# แปลงชื่อเดือนเป็นหมายเลขเดือน
map_month = df['month'].map(month_mapping)
map_month

# คำนวณค่าเฉลี่ยของ 'Chlo_b หรือ ความเข้มข้นของคลอโรฟิลล์ b' สำหรับแต่ละเดือนและแต่ละสถานี
mean_chlo_b_per_month_station = df.groupby(['month', 'station'])['Chlo_b'].mean().unstack()

# สร้างคอลัมม์ month แยก เพื่อนำไป sort
index_df = pd.DataFrame(mean_chlo_b_per_month_station.index, columns=['month'])

# sort ข้อความโดยการแปลงตัวอักษรเป็นตัวเลข
index_df['month_num'] = index_df['month'].map(month_mapping)
index_df = index_df.sort_values(by='month_num')

# แสดงผลโดยเรียงเดือนจาก index_df
mean_chlo_b_per_month_station.index = index_df['month']
mean_chlo_b_per_month_station
```
