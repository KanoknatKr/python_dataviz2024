````
movie_directors = movie_df.groupby('Director').agg({
    'Name': 'count',  # นับจำนวนภาพยนตร์ที่กำกับโดยผู้กำกับแต่ละคน
    'Rating': 'mean', # คำนวณคะแนนเฉลี่ยของภาพยนตร์ที่กำกับโดยผู้กำกับแต่ละคน
    'Votes': 'sum'   # รวมจำนวนโหวตของภาพยนตร์ที่กำกับโดยผู้กำกับแต่ละคน
}).reset_index()
````

````
movie_directors
````

````
# แปลงคอลัมน์ 'Duration' เป็นสตริง
movie_df['Duration'] = movie_df['Duration'].astype(str) \
.str.replace(' min', '') \
.astype(int)
````

````
# คำนวณอันดับของผู้กำกับโดยพิจารณาจากจำนวนภาพยนตร์ คะแนนเฉลี่ย และจำนวนโหวต
movie_directors['Rank'] = movie_directors[['Directed','Rating', 'Votes']].rank(method='max', ascending=False).mean(axis=1)
# เรียงลำดับผู้กำกับตามอันดับและเลือก 3 อันดับแรก
top_directors = movie_directors.sort_values(by='Rank').head(3)
# แสดงผลลัพธ์
print(top_directors)
````

````
# คำนวณอันดับของผู้กำกับโดยพิจารณาจากจำนวนภาพยนตร์ คะแนนเฉลี่ย และจำนวนโหวต
movie_directors['Rank'] = movie_directors[['Name','Rating', 'Votes']].rank(method='max', ascending=False).mean(axis=1) # Use 'Name' instead of 'Directed'
# เรียงลำดับผู้กำกับตามอันดับและเลือก 3 อันดับแรก
top_directors = movie_directors.sort_values(by='Rank').head(3)
# แสดงผลลัพธ์
print(top_directors)
````

````
# สร้าง list ของผู้กำกับ 3 อันดับแรก
top_directors_list = top_directors['Director'].tolist()

# วนลูปผ่านรายชื่อผู้กำกับ 3 อันดับแรก
for director in top_directors_list:
    # กรองข้อมูลใน DataFrame movie_df เพื่อเลือกเฉพาะข้อมูลของผู้กำกับปัจจุบัน
    director_info = movie_df[movie_df['Director'] == director]
    # กำหนดค่าตัวแปร director ให้เป็นชื่อของผู้กำกับปัจจุบัน
    director = director_info['Director'].iloc[0]
    # Check if 'Title' column exists before accessing it
    if 'Title' in director_info.columns: # Removed extra indent here
        title = director_info['Title'].iloc[0]
        print(f"Title of one of their films: {title}") # Print the title if it exists
    year = director_info['Year'].iloc[0]
    print(f"ผู้กำกับ: {director}")
    print(f"ผลงานการกำกับภาพยนต์:")
    print(f"- ผลงานภาพยนต์ที่สร้าง: {director_info['Name'].count()} เรื่อง")
    print(f"- คะแนนเฉลี่ยรวม: {director_info['Rating'].mean():.2f}")
    print(f"- จำนวนผลโหวตรวม: {director_info['Votes'].sum()}")
    print()
````
